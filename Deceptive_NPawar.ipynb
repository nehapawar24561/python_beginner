# Data Analysis and Conditioning:

```python
import pandas as pd
import numpy as np
```

```python
#importing White House salaries - Obama admin.csv file
obama_Staff = pd.read_csv('White House salaries - Obama admin.csv')
```

```python
#importing White House salaries - Trump admin.csv file
trump_Staff = pd.read_csv('White House salaries - Trump admin.csv')
```

```python
#original Obama file
obama_Staff.head()
```

```python
#Original Trump file
trump_Staff.head()
```

```python
#getting rid of $ and , from the salary number from Obama dataframe
obama_Staff['SALARY'] = obama_Staff['SALARY'].str.replace('$', '')
obama_Staff['SALARY'] = obama_Staff['SALARY'].str.replace(',', '')
```

```python
#typecasting SALARY as float
obama_Staff['SALARY'] = obama_Staff['SALARY'].astype(float)
```

```python
#Dropping unwanted columns from Obama dataframe
obama_Staff.drop(['NAME','POSITION TITLE', 'STATUS', 'PAY BASIS'],axis=1,inplace=True)
```

```python
#Creating SALARY buckets for Obama dataframe
obama_Staff['Staff_Salary_Buckets'] = pd.cut(obama_Staff['SALARY'], [-1, 30000, 60000, 90000,120000, 150000, 
                                                                          180000, 210000],
                                                 labels = ['30000', '60000', '90000','120000', '150000', 
                                                                          '180000', '210000'])
```

```python
#grouping by SALARY buckets to get count of employees, mean salary and total salary for each SALARY bucket
obama_Staff1 = obama_Staff.groupby('Staff_Salary_Buckets', as_index = False)['SALARY'].agg(['count',
                                                                                       'mean','sum']).rename(columns=
                                                    {'Staff_Salary_Buckets' : 'Staff_Salary_Buckets',
                                                     'count' : 'Count_of_employees_Obama',
                                                     'mean' : 'Salary_Mean_Obama',
                                                     'sum' :  'Total_Salary_Bucket_Obama'}).reset_index()
```

```python
#Please refer writeup and references for explanations
Number_of_laws_signed_by_Obama_in_3_months = 14
Number_of_total_pages_for_these_laws_Obama = 1602
```

```python
# 12 = number of months in year, 3 = 3 months because metric is defined for 3 months
obama_Staff1['Money_spent_per_law_100days_Obama'] = (obama_Staff1.Salary_Mean_Obama * obama_Staff1.Count_of_employees_Obama*3)/(12*Number_of_laws_signed_by_Obama_in_3_months*Number_of_total_pages_for_these_laws_Obama)
```

```python
#getting rid of $ and , from the salary number from Trump dataframe
trump_Staff['SALARY'] = trump_Staff['SALARY'].str.replace('$', '')
trump_Staff['SALARY'] = trump_Staff['SALARY'].str.replace(',', '')
```

```python
#typecasting SALARY as float
trump_Staff['SALARY'] = trump_Staff['SALARY'].astype(float)
```

```python
#dropping unwanted columns
trump_Staff.drop(['NAME','POSITION TITLE', 'STATUS', 'PAY BASIS'],axis=1,inplace=True)
```

```python
#Creating SALARY buckets for Trump dataframe
trump_Staff['Staff_Salary_Buckets'] = pd.cut(trump_Staff['SALARY'], [-1, 30000, 60000, 90000,120000, 150000, 
                                                                          180000, 210000]
                                                 ,
                                                 labels = ['30000', '60000', '90000','120000', '150000', 
                                                                          '180000', '210000'])
```

```python
#grouping by SALARY buckets to get count of employees, mean salary and total salary for each SALARY bucket
Trump_staff1 = trump_Staff.groupby('Staff_Salary_Buckets', as_index = False)['SALARY'].agg(['count',
                                                                                       'mean', 'sum']).rename(columns=
                                                    {'Staff_Salary_Buckets' : 'Staff_Salary_Buckets',
                                                     'count' : 'Count_of_employees_Trump',
                                                     'mean' : 'Salary_Mean_Trump',
                                                     'sum' : 'Total_Salary_Bucket_Trump'}).reset_index()
```

```python
#Please refer writeup and references for explanations
Number_of_laws_signed_by_Trump_in_3_months = 29
Number_of_total_pages_for_these_laws_Trump = 133
```

```python
# 12 = number of months in year, 3 = 3 months because metric is defined for 3 months
Trump_staff1['Money_spent_per_law_100days_Trump'] = (Trump_staff1.Salary_Mean_Trump * 
                                                     Trump_staff1.Count_of_employees_Trump*3)/(12 * Number_of_laws_signed_by_Trump_in_3_months * Number_of_total_pages_for_these_laws_Trump)
```

```python
#merging both the datasets to get combined file for visualizations
white_house4= obama_Staff1.merge(Trump_staff1, on='Staff_Salary_Buckets')
```

```python
#adding a column to get difference between employee count for Obama and Trump
white_house4['Obama_vs_Trump_Staff_Count'] = white_house4.Count_of_employees_Obama - white_house4.Count_of_employees_Trump
```

```python
#coverting dataframe into csv file
white_house4.to_csv('white_house_with_final_ratios.csv')
```
