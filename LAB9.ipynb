```python
#importing .csv file
LAB3_ticker = pd.read_csv('LAB3_ticker.csv')
```

```python
#updating column name to Ticker_symbol to merge files 
ticker_data_LAB3.columns = ['Ticker_symbol', 'Date','Price' ]
```

```python
#only selecting records with Entity Home Depot, NASDAQ,UPS, Vodafone, Yahoo
LAB3_ticker1= LAB3_ticker.loc[LAB3_ticker['Ticker_symbol'].isin(['VOD', 'AABA', 'UPS', 'NDAQ', 'HD'])]
```

```python
#merging files with only selected entities data
Complete_data = LAB3_ticker1.merge(ticker_data_LAB3, on='Ticker_symbol')
```
