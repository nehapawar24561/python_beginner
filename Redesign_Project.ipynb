```python
#mapping coverted data to phone dataset
phone.phone_brand = phone.phone_brand.map(pd.Series(english_phone_brands_mapping), na_action='ignore')
```

```python
#reset index to simplify merge operation
gatrain= gatrain.reset_index()
phone = phone.reset_index()
```

```python
#merging gatrain and phone dataframes
train = gatrain.merge(phone[['device_id','phone_brand','device_model']], how='left',on='device_id')
```

```python
#checking null records
NullRows = train[train.isnull().any(axis=1)]
```

```python
#skipping rows with null phone_brand entry
train1 = train.dropna(subset=['phone_brand'], how='any')
```

```python
#this is just to get idea about the data
# Top brands by number of devices in the market
traingroup = train1.groupby('phone_brand',as_index=False).agg({'device_id': 'count'}).rename(columns=
                                                    {'device_id' : 'count_device_id'})
topbyusers = traingroup.sort_values(by = ['count_device_id'], ascending = False)
```

```python
#this is just to get idea about the data
#Top groups and phone brands as per number of devices in the market
topagegroup = train1.groupby(['group', 'phone_brand'],as_index=False).agg({'device_id': 'count'}).rename(columns=
            {'device_id' : 'count_device_id'}).sort_values(by = ['group','count_device_id'], ascending = [True, False])
```

```python
#creating dummy columns for Gender Male and Female
train5 = pd.get_dummies(train1, columns = ['gender'])
```

```python
#creating a new dataframe which shows 
train6 = train5.groupby(['phone_brand'], as_index = False).agg({'gender_F': 'sum', 
                                     'gender_M': 'sum',
                                    'device_id': 'count',
                                     'age': 'mean'}).rename(columns=
                                                    {'phone_brand' : 'Phone Brand',
                                                     'gender_F' : 'Female Consumer Count',
                                                     'gender_M' : 'Male Consumer Count',
                                                     'age' : 'Average Age',
                                                      'device_id': 'Total Consumer Count'}).sort_values(by = ['Female Consumer Count', 'Male Consumer Count'], 
                                                                     ascending = [False, False])
```

```python
#correcting column order of the dataframe
train6 = train6[['Phone Brand', 'Average Age','Total Consumer Count','Female Consumer Count', 'Male Consumer Count']]
```

```python
#creating column which signifies ratio of female consumer count to male consumer count per brand
train6['F_gender_ratio'] = train6['Female Consumer Count']/train6['Male Consumer Count']
#taking care of infinty values arised due to ratio calculation
train6.loc[~np.isfinite(train6['F_gender_ratio']), 'F_gender_ratio'] = np.nan 
```

```python
#creating column which signifies ratio of male consumer count to female consumer count per brand
train6['M_gender_ratio'] = 1 / train6['F_gender_ratio']
#taking care of infinty values arised due to ratio calculation
train6.loc[~np.isfinite(train6['M_gender_ratio']), 'M_gender_ratio'] = np.nan 
```

```python
#convering column names into upper case
train6.columns = ['PHONE_BRAND', 'AVERAGE AGE', 'TOTAL CONSUMER COUNT', 'FEMALE CONSUMER COUNT','MALE CONSUMER COUNT','F_GENDER_RATIO', 'M_GENDER_RATIO']
```

```python
#converting phone brand names into upper case
train6['PHONE_BRAND'] = train6.PHONE_BRAND.str.upper()
```

```python
#coverting dataframe into csv file
train6.to_csv('gender_ratio.csv')
```
